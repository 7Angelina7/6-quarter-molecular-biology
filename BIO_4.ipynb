Alpha-2-Macroglobulin
Большой белок плазмы, обнаруженный в крови. Он в основном вырабатывается печенью, а также локально синтезируется макрофагами, фибробластами и клетками коры надпочечников. У человека он кодируется геном A2M.
Белок контролирует развитие инфекций и воспалительных процессов.

В структуре А (SWISS MODEL) и в структуре Б (ALPHA FOLD) прослеживается характерное начало структуры в виде хвостика. Далее обе конфигурации напоминают ушную раковину.
Различия в структурах все же есть.
